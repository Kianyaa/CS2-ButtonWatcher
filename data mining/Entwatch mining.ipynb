```python
cwd = os.getcwd()

if "entwatch" in os.listdir(cwd):
    entwatch_path = os.path.join(cwd, "entwatch")
    print(entwatch_path)

path = entwatch_path
filtered_all = []
count = 0

with open('output.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(['item_name'])  # header


for file in os.listdir(path):
    if file.endswith('.jsonc'):
        print(file, ':', end='\n\n')
        
        with open(path+ '\\' + file, 'r') as file:
            data = json.load(file)
            
        item_split = []
        item_unique = []
        for each_item in data:
            print(each_item['name'].split(' '))
            item = each_item['name'].split(' ')
            item_split.append(item)

        print()
        
        for item in item_split:
            for item_split in item:
                item_unique.append(item_split)

        to_insert = ['the', 'of', 'materia', '1', '2'] # Lower_Case only!!!
        
        for item in to_insert:
                item_unique.insert(0, item)
        
        item_unique = list(set(item_unique))

        filtered = [word.lower() for word in item_unique if word.lower() not in to_insert]

        filtered_all.extend(filtered)

        
        count += 1
        
        if count == 116:
            break

        with open('output.csv', 'a', newline='') as f:
            writer = csv.writer(f)  # header
            for item in filtered:
                writer.writerow([item])
```

```python
# Load the CSV into a DataFrame
df = pd.read_csv('output.csv')

# Convert the column to a list, remove duplicates using set, and save back to CSV
unique_items = list(set(df['item_name']))

# Create a dictionary with the key 'item_name' and the list as its value
data_dict = {'item_name': unique_items}

# Convert the dictionary into a JSON format
json_data = json.dumps(data_dict, indent=4)

# Add comments manually in JSONC format (JSON does not support comments natively, but JSONC allows them)
jsonc_data = f"// This is a list of items name updated 6/04/2025\n{json_data}"

# Save it to a JSONC file
file_path = 'data_item_names.jsonc'
with open(file_path, 'w') as file:
    file.write(jsonc_data)

print('completed export',file_path)  # Return the file path as confirmation
```
